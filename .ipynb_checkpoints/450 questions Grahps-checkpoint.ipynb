#### BFS Traversal

```python
class Graph:
    def __init__(self,vertices,is_directed=False):
        self.is_directed = is_directed
        self.vertices = vertices
        self.adj_list = {}
        
        for v in self.vertices:
            self.adj_list[v] = []
            
            
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
    def bfs(self, source):
        hash_map = {}
        
        queue = []
        
        out = []
        
        hash_map[source] = True
        queue.append(source)
            
        while(queue):
            source = queue.pop(0)
            out.append(source)
            
            
            for i in self.adj_list[source]:
                if(hash_map.get(i) == None):
                    queue.append(i)
                    hash_map[i] = True
                    
        print(out)
            
        
    def print(self):
        for i in self.vertices:
            print(i,"->",self.adj_list[i])
            

            
vertics = [0,1,2,3,4]
edges = [(0,1),(0,2),(0,3),(2,4)]
g = Graph(vertics,True)
for v,e in edges:
    g.add_edge(v,e)
    
g.print()
print("bfs")
g.bfs(0)
```

#### DFS (Recursion) 

```python
class Graph:
    hash_map = {}
    dfs_out = []
    def __init__(self,vertices,is_directed=False):
        self.is_directed = is_directed
        self.vertices = vertices
        self.adj_list = {}
        
        for v in self.vertices:
            self.adj_list[v] = []
            
            
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
            
    def dfs(self, source):
        Graph.dfs_out.append(source)
        Graph.hash_map[source]=  True
        
        for i in self.adj_list[source]:
            if(Graph.hash_map.get(i) == None):
                self.dfs(i)
                
                
#     in run DFS from all unvisited nodes after a DFS.The recursive function remains the same.

    def dfs_disconnected_scenario(self, v_array):
        for i in v_array:
            if(self.hash_map.get(i)==None):
                self.dfs(i)
           
    def print(self):
        for i in self.vertices:
            print(i,"->",self.adj_list[i])
            
            
g = Graph
vertics = [0,1,2,3,4]
edges = [(0,1),(2,1),(2,3),(2,4),(3,4),(4,0),(4,2)]
g = Graph(vertics,True)
for v,e in edges:
    g.add_edge(v,e)
    
g.print()

print("\n")
print("Normal DFS funtion with a source node")
g.dfs(0)
print(g.dfs_out)
print("\n")

print("Disconnected graph which does the traversal from all the nodes")
g.dfs_disconnected_scenario(vertics)
print(g.dfs_out)
```

#### DFS and BFS Traversal (similar pattern - Stack and Queue)

```python
class Graph:
    hash_map = {}
    dfs_out = []
    def __init__(self,vertices,is_directed=False):
        self.is_directed = is_directed
        self.vertices = vertices
        self.adj_list = {}
        
        for v in self.vertices:
            self.adj_list[v] = [] 
            
            
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
    
    def bfsOfGraph(self, V, adj):
        queue = []
        hash_map = {}
        out = []
        
        queue.append(0)
        
        
        while(queue):
            popped = queue.pop(0)
            if(hash_map.get(popped) == None):
                out.append(popped)
            hash_map[popped] = True
            print(hash_map)
            
            for i in adj[popped]:
                if(hash_map.get(i) == None):
                    queue.append(i)
        
        return out

    
    def dfsOfGraph(self, V, adj):
        stack = []
        hash_map = {}
        out = []
        print(adj)
        
        stack.append(0)
        
        while(stack):
            popped = stack.pop()
            if(hash_map.get(popped) == None):
#                 print(out)
                out.append(popped)
            
            hash_map[popped]  = True
            print(hash_map)
              
            for i in range(len(adj[popped])-1, -1, -1):
                if(hash_map.get(adj[popped][i]) == None):
                    stack.append(adj[popped][i])
                    
        return out
        
            
    def print(self):
        for i in self.vertices:
            print(i,"->",self.adj_list[i])
            
            
g = Graph
vertics = [0,1,2,3,4]
edges = [(0,1),(1,0),(1,2),(2,1),(2,3),(2,4),(3,4),(4,0),(4,2)]
g = Graph(vertics,True)
for v,e in edges:
    g.add_edge(v,e)
    
g.print()

print("\n")
print("Normal BFS funtion with a source node")  
print(g.bfsOfGraph(0,g.adj_list))
# print(g.dfs_out)
print("\n")

print("\n")
print("Normal DFS funtion with a source node")
print(g.dfsOfGraph(0,g.adj_list))
# print(g.dfs_out)
print("\n")
```

#### Detect Cycle in Directed Graph (DFS algorithm - White Set, Black Set,  Grey Set). 

```python
class Graph:
    def __init__(self, vertices, is_directed = False):
        self.is_directed = is_directed
        self.vertices = vertices
        self.adj_list = {}
        
        self.white_set = {}
        self.grey_set = {}
        self.black_set = {} 
        
        
        for i in self.vertices:
            self.adj_list[i] = []

            
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
    def print(self):
        for i in self.adj_list:
            print(i, "-->", self.adj_list[i])
            
    def detect_cycle(self):
        for i in self.adj_list:
            print(i)
            self.white_set[i] = True
            
        while(len(self.white_set) > 0):
            curr = next(iter(self.white_set))
            if(self.dfs(curr)):
                return True 
            
        return False

    def dfs(self, vertex):
        print("form white to grey")
        self.move_vertex(vertex, self.white_set, self.grey_set)
        
        for i in self.adj_list[vertex]:
            if(i in self.black_set):
                continue
            if(i in self.grey_set):
                return True
            
            if( self.dfs(i) == True):
                return True
            
        print("from grey to black")
        self.move_vertex(vertex, self.grey_set, self.black_set)
        return False
    
    def move_vertex(self,vertex, source_set, destination_set):
        print("v---------->",vertex)
        print("Src--------->", source_set)
        print("dest ------->",destination_set)
        source_set.pop(vertex)
        destination_set[vertex] = True
        
    
vertices = [0,1,2,3]
edges = [(2,0),(0,1), (1,3), (3,1), (1,0)]

g = Graph(vertices, True)

for v,e in edges:
    g.add_edge(v,e)
    
g.print()
g.detect_cycle()
        
```

#### Topological Sorting BFS -Kahn's Algo (in order to understand Cycle detection -->BFS)

```python
class Graph:
    def __init__(self, vertices, is_directed):
        self.vertices = vertices
        self.is_directed = is_directed
        self.adj_list = {}
        
        for i in vertices:
            self.adj_list[i] = []
            
    
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
    def print(self):
        for i in self.adj_list:
            print(i,"-->",self.adj_list[i])
            
    def topological_sort(self):
        inDegree = [0]*len(self.vertices)
        topoSort = []
        queue = []
        
        for i in self.vertices:
            for j in self.adj_list[i]:
                inDegree[j] +=1
                
        for i in range(len(inDegree)):
            if(inDegree[i] == 0):
                queue.append(i)
                
        while(len(queue)!= 0):
            popped = queue.pop(0)
            topoSort.append(popped)
            for i in self.adj_list[popped]:
                inDegree[i] = inDegree[i] - 1
                if(inDegree[i] == 0):
                    queue.append(i)
        
        return topoSort
    

vertices = [0,1,2,3,4,5]
edges = [(2,3), (3,1), (4,0),(4,1),(5,0),(5,2)]

g = Graph(vertices,True)
for v,e in edges:
    g.add_edge(v,e)
g.print()

print(g.topological_sort())
            
        
```

#### Rat in a maze problem 

```python
class Solution:
    def __init__(self):
        self.out = []
        
        
    def maze(self,i, j, route, m, a):
#         print(i,j)
#         print("------------",route)
        if(a[i][j] == 0):
            return 
        a[i][j] = 0
        
        if(i==m and j == m):
            self.out.append(route)
        
        #down 
        if(i<m and j<=m and a[i+1][j] == 1):
            self.maze(i+1, j, route+"d", m, a)
        #left    
        if(i<=m and j>0 and a[i][j-1] == 1):
            self.maze(i, j-1, route+"l", m, a)
        #right    
        if(i<=m and j<m and a[i][j+1] == 1):
            self.maze(i, j+1, route+"r", m, a)
        #up
        if(i>0 and j<=m and a[i-1][j] == 1):
            self.maze(i-1, j, route+"u", m, a)
        
        a[i][j] = 1
        
a = [[1, 0, 0, 0],
         [1, 1, 0, 1], 
         [1, 1, 0, 0],
         [0, 1, 1, 1]]

s = Solution()
print(s.out)
m = len(a[0])-1
s.maze(0,0,"",m,a)
print(s.out)
        
```
