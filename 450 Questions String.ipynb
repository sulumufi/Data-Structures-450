#### Longest common Subsecquence (Bottom Up) [Dp]

```python
def longest_common(s1,s2):
    len_s1 = len(s1)
    len_s2 = len(s2)
    
    table = [[0 for i in range(len_s2+1)] for i in range(len_s1+1)]
        
    for i in range(len_s1 + 1):
        for j in range(len_s2 + 1):
            if(i == 0 or j == 0):
                table[i][j] = 0
            elif(s1[i-1] == s2[j-1]):
                table[i][j] = 1 + table[i-1][j-1]
            elif(s1[i-1] != s2[j-1]):
                table[i][j] = max(table[i][j-1],table[i-1][j])
                
    return table[-1][-1]
        
        
s1 = "GXTXAYB"
s2 ="AGGTAB"

print(longest_common(s1,s2))

                
                
    
```

#### Longest Common Subsequence [Bottom up]

```python
def lcs(s1,s2):
    if(len(s1) == 0 or len(s2) == 0):
        return 0
    last_of_s1 = s1[-1]
    last_of_s2 = s2[-1]
    
    s1_without_last = s1[:-1]
    s2_without_last = s2[:-1]
    
    if(last_of_s1 == last_of_s2):
        return 1 + lcs(s1_without_last, s2_without_last)
    else:
        return max(lcs(s1,s2_without_last),lcs(s1_without_last,s2))
    
s1 = "aggtab"
s2 = "gxtxayb"

print(lcs(s1,s2))
```

```python
# Following program is the python implementation of 
# Rabin Karp Algorithm given in CLRS book 
  
# d is the number of characters in the input alphabet 
d = 256
  
# pat  -> pattern 
# txt  -> text 
# q    -> A prime number 
  
def search(pat, txt, q): 
    M = len(pat) 
    N = len(txt) 
    i = 0
    j = 0
    p = 0    # hash value for pattern 
    t = 0    # hash value for txt 
    h = 1
  
    # The value of h would be "pow(d, M-1)%q" 
    for i in range(M-1): 
        h = (h*d)%q 
  
    # Calculate the hash value of pattern and first window 
    # of text 
    for i in range(M): 
        p = (d*p + ord(pat[i]))%q 
        t = (d*t + ord(txt[i]))%q 
  
    # Slide the pattern over text one by one 
    for i in xrange(N-M+1): 
        # Check the hash values of current window of text and 
        # pattern if the hash values match then only check 
        # for characters on by one 
        if p==t: 
            # Check for characters one by one 
            for j in range(M): 
                if txt[i+j] != pat[j]: 
                    break
                else: j+=1
  
            # if p == t and pat[0...M-1] = txt[i, i+1, ...i+M-1] 
            if j==M: 
                print("Pattern found at index " + str(i)) 
  
        # Calculate hash value for next window of text: Remove 
        # leading digit, add trailing digit 
        if i < N-M: 
            t = (d*(t-ord(txt[i])*h) + ord(txt[i+M]))%q 
  
            # We might get negative values of t, converting it to 
            # positive 
            if t < 0: 
                t = t+q 
  
# Driver Code 
txt = "GEEKS FOR GEEKS"
pat = "GEEK"
  
# A prime number 
q = 101 
  
# Function Call 
search(pat,txt,q) 
  
# This code is contributed by Bhavya Jain 
```

#### Find Minimum Charachters to be added at the front to make the string Palindrome (KMP - Lps_array)

```python
def lps(s):
    n = len(s)
    print(s)
    p = [0] * n
    
    i = 0
    j = 1
    
    while(j<n):
        if(s[i] == s[j]):
            i = i + 1
            p[j] = i
            j = j + 1
        else:
            if(i != 0):
                i = p[i-1]
            else:
                j = j + 1
                
    return p[n-1]


def add_char_to_palidrome(s):
    n = len(s)
    s1 = s[::-1]
    
    new_str = s + "*" + s1
    
    index_found = lps(new_str)
    
    return n - index_found



a = "aacecaaaa"
add_char_to_palidrome(a)

                
                
```

#### Remove Consecutive duplicate. (GFG)

```python
def removeConsecutiveCharacter(a):
        stack = []
        for i in range(len(a)):
            if(len(stack) == 0):
                stack.append(a[i])
            else:
                if(stack[-1] != a[i]):
                    stack.append(a[i])
        stack = "".join(map(str,stack))            
        return stack
    
a = "abbaca"
removeConsecutiveCharacter(a)
```

```python
def pateren_astrick(s,p):
    dp = [[0 for i in range(len(p)+1)] for i in range(len(s)+1)]
    
    m = len(s)
    n = len(p)
    dp[0][0] = True
    for i in range(1,m+1):
        dp[i][0] = False
        
    
    for j in range(1,n+1):
        if(p[j-1] == "*"):
            dp[0][j] = dp[0][j-1]
        else:
            dp[0][j] = False
            
    for i in range(len(dp)):
        print(dp[i])
            
    for i in range(1,m+1):
        for j in range(1,n+1):
            if(s[i-1] == p[j-1] or p[j-1] == "?"):
                dp[i][j] = dp[i-1][j-1]
            elif(p[j-1] == "*"):
                dp[i][j] = dp[i][j-1] or dp[i-1][j]
            else:
                dp[i][j] = False
    for i in range(len(dp)):
        print(dp[i])
    return dp[-1][-1]

s = "albmnc"
pattern = "**a?b*c"
pateren_astrick(s,pattern)
```
