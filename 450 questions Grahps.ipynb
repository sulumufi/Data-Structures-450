#### Topological Sorting BFS -Kahn's Algo (in order to understand Cycle detection -->BFS)

```python
class Graph:
    def __init__(self, vertices, is_directed):
        self.vertices = vertices
        self.is_directed = is_directed
        self.adj_list = {}
        
        for i in vertices:
            self.adj_list[i] = []
            
    
    def add_edge(self, vertex, edge):
        self.adj_list[vertex].append(edge)
        if(self.is_directed == False):
            self.adj_list[edge].append(vertex)
            
    def print(self):
        for i in self.adj_list:
            print(i,"-->",self.adj_list[i])
            
    def topological_sort(self):
        inDegree = [0]*len(self.vertices)
        topoSort = []
        queue = []
        
        for i in self.vertices:
            for j in self.adj_list[i]:
                inDegree[j] +=1
                
        for i in range(len(inDegree)):
            if(inDegree[i] == 0):
                queue.append(i)
                
        while(len(queue)!= 0):
            popped = queue.pop(0)
            topoSort.append(popped)
            for i in self.adj_list[popped]:
                inDegree[i] = inDegree[i] - 1
                if(inDegree[i] == 0):
                    queue.append(i)
        
        return topoSort
    

vertices = [0,1,2,3,4,5]
edges = [(2,3), (3,1), (4,0),(4,1),(5,0),(5,2)]

g = Graph(vertices,True)
for v,e in edges:
    g.add_edge(v,e)
g.print()

print(g.topological_sort())
            
        
```

##### Kirchoff's Theorem - Total number of spanning trees in a graph 

```python
from typing import List

class Solution:
    def findDeterminant(self, Matrix: List[List[int]]) -> int:
        print(Matrix)
        # Base cases
        if len(Matrix) == 1:
            return Matrix[0][0]
        elif len(Matrix) == 2:
            return Matrix[0][0] * Matrix[1][1] - Matrix[0][1] * Matrix[1][0]

        # Recursive case
        det = 0
        for p in range(len(Matrix)):
            # Creating the submatrix by excluding the current row and column
            tempMatrix = []
            for i in range(1, len(Matrix)):
                row = [Matrix[i][j] for j in range(len(Matrix)) if j != p]
                if row:
                    tempMatrix.append(row)

            # Cofactor expansion along the first row
            det += Matrix[0][p] * (-1) ** p * self.findDeterminant(tempMatrix)
        
        return det

    def countSpanningTrees(self, graph: List[List[int]], n: int, m: int) -> int:
        # Initialize adjacency matrix
        adjMatrix = [[0] * n for _ in range(n)]
        for edge in graph:
            x, y = edge
            adjMatrix[x][y] = 1
            adjMatrix[y][x] = 1

        # Calculate degrees
        degree = [0] * n
        for i in range(n):
            for j in range(n):
                if adjMatrix[i][j] == 1:
                    degree[i] += 1

        # Construct Laplacian matrix
        for i in range(n):
            adjMatrix[i][i] = degree[i]
        for i in range(n):
            for j in range(n):
                if i != j and adjMatrix[i][j] == 1:
                    adjMatrix[i][j] = -1

        # Create the submatrix by excluding the first row and first column
        submatrix = [[adjMatrix[i][j] for j in range(1, n)] for i in range(1, n)]

        # Return the determinant of the submatrix
        return self.findDeterminant(submatrix)


# Number of vertices 
n = 6
m = 5

edges = [[0, 3], [0, 1], [1, 2], [1, 5], [3, 4]]
s = Solution()
s.countSpanningTrees(edges,n,m) 
```

```python
import sys

class Solution:
    def isNegativeWeightCycle(self, n, edges):
        #Code here
#         print(edges)
        
        weight_array = [sys.maxsize] * n;
        weight_array[0] = 0
        
        last_iteration_weight = [0]*n;
        
        iteration = 0
        while(iteration < n+1):
            
            for i,j,w in edges:
                if(weight_array[i] + w < weight_array[j]):
                    weight_array[j] = weight_array[i] + w
            
            if(iteration == n-1):
                for k in range(n):
                    last_iteration_weight[k] = weight_array[k]
                    
            iteration = iteration + 1
        
        for i in range(0,n):
            if(last_iteration_weight[i] != weight_array[i]):
                return 1
        return 0
```

#### Water Jug problem DFS

```python
def water_jug_dfs(C1, C2, T):
    from collections import deque

    # Helper function to pour water between jugs
    def pour(x, y, Cx, Cy):
        transfer = min(x, Cy - y)  # Transfer as much as possible without spilling
        return x - transfer, y + transfer

    # DFS function
    def dfs(state, visited, path):
        x, y = state
        if (x, y) in visited:  # If already visited, skip
            return False
        if x == T or y == T:  # If the target is achieved
            path.append((x, y))
            return True

        visited.add((x, y))  # Mark the current state as visited
        path.append((x, y))

        # Generate all possible next states
        next_states = [
            (C1, y),           # Fill jug1
            (x, C2),           # Fill jug2
            (0, y),            # Empty jug1
            (x, 0),            # Empty jug2
            pour(x, y, C1, C2),  # Pour jug1 -> jug2
            pour(y, x, C2, C1)[::-1],  # Pour jug2 -> jug1
        ]

        # Explore all next states
        for next_state in next_states:
            if dfs(next_state, visited, path):
                return True

        path.pop()  # Backtrack
        return False

    # Initialize
    visited = set()
    path = []

    # Start DFS
    initial_state = (0, 0)
    if dfs(initial_state, visited, path):
        print("Solution path:")
        for step in path:
            print(step)
    else:
        print("No solution exists.")

# Example usage
water_jug_dfs(C1=4, C2=3, T=2)
```

```python
#User function Template for python3
from typing import List
class Solution:
    def minCashFlow(self, transactions: List[List[int]], N: int) -> List[List[int]]:
        def getMaxIndex(arr):
            max_index = 0
            for i in range(1, len(arr)):
                if arr[i] > arr[max_index]:
                    max_index = i
            return max_index

        # Helper function to get the index of the person with maximum debit
        def getMinIndex(arr):
            min_index = 0
            for i in range(1, len(arr)):
                if arr[i] < arr[min_index]:
                    min_index = i
            return min_index
            
        # Helper function to minimize cash flow recursively
        def minimizeCashFlowRec(amount):
            max_credit = getMaxIndex(amount)
            max_debit = getMinIndex(amount)

            # If all amounts are settled
            if amount[max_credit] == 0 and amount[max_debit] == 0:
                return

            # Find the minimum of max credit and max debit
            settlement = min(amount[max_credit], -amount[max_debit])

            # Settle the amounts
            amount[max_credit] -= settlement
            amount[max_debit] += settlement

            # Log the transaction
            result[max_debit][max_credit] += settlement

            # Recur for remaining unsettled amounts
            minimizeCashFlowRec(amount)
            
            # Step 1: Calculate net amount each friend owes or is owed
        net_amount = [0] * N
        for i in range(N):
            for j in range(N):
                net_amount[i] += transactions[j][i] - transactions[i][j]

        # Initialize the result transactions array
        result = [[0 for _ in range(N)] for _ in range(N)]

        # Step 2: Minimize cash flow using recursion
        minimizeCashFlowRec(net_amount)
        
        return result

            
            
```
