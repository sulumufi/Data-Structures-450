#### Linked List Basics

```python
class Node:
    def __init__(self, data = None, next = None ):
        self.data = data
        self.next = next
        
class LinkedList :
    def __init__(self):
        self.head = None
    
    def insert_at_beginning(self, data):
        node = Node(data, self.head)
        self.head = node
        
        
    def insert_at_end(self, data):
        if(self.head == None):
            node = Node(data, None)
            self.head = node
            return
            
        i = self.head
        
        while(i.next != None):
            i = i.next
        i.next  = Node(data, None)
        
    
    def print(self):
        if self.head is None:
            print("LinkedList is Empty")
            return 
            
        i = self.head
        llstr = ""
        
        while(i):
            llstr += str(i.data) + "--->"
            i = i.next
            
        print(llstr)
        
    def get_lenght(self):
        count = 0
        i = self.head
        while(i.next):
            count = count + 1
            i = i.next
        return count 
    
    def del_at_index(self, index):
        if(index<0 or index > self.get_lenght()):
            raise Exception("invalid index")
            
        if(index == 0):
            self.head = self.head.next
        
        count = 0 
        i = self.head
        
        while(i.next):
            if(count == index - 1):
                i.next = i.next.next
                break
            i = i.next
            count = count + 1
    
    def insert_at_index(self, index, data):
        if(index <0 or index >self.get_lenght()):
            raise Exception("invalid index")
            
        if(index == 0):
            insert_at_beginning(data)
            return 
        
        count = 0 
        i = self.head

        while(i.next):
            if(count == index-1):
                node = Node(data, i.next)
                i.next = node
                break
            count = count + 1
            i = i.next
            
        
        
        
        
    def inset_data_list(self, data_list):
        for i in data_list:
            self.insert_at_end(i)
            
    def using_hash(self):
        hashmap = {}
        
        i = self.head
        
        while(i):
            hashmap[i] = True
            i = i.next
            
        print(hashmap)
        
        
             
    
        
         
if __name__ == "__main__":
    ll = LinkedList()
    ll.inset_data_list([1,2,3,4,5,6,7,8,9])
#     ll.insert_at_index(2, "im_at_index_2")
    ll.print()
    ll.using_hash()
    
        
        
    
    
        
    
```

#### CLone Linked List (O(n) space and O(n) Time)

```python
class Node:
    def __init__(self,data = None , next=None, arb= None):
        self.data  = data 
        self.next = next
        self.arb = arb
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def add_at_end(self, data):
        if(self.head == None):
            self.head = Node(data)
        else:
            i = self.head 
            
            while(i.next):
                i = i.next
            i.next = Node(data)
            
    def insert_data_list(self, data_list):
        for i in range(len(data_list)):
            self.add_at_end(data_list[i])
            
            
    def set_arb(self, head, a, b):
        h = head
        i = 1
        
        while(i<a and h):
            h = h.next
            i = i + 1
            
        an = h
        
        h = head
        i = 1
        
        while(i<b and h):
            h = h.next
            i = i + 1
            
        if(an):
            an.arb = h

            
    def print(self):
        if(self.head == None):
            print("Linked List is empty")
            return 
        
        i = self.head
        str1 = ""
        while(i):
            str1 = str1 + str(i.data)+ "--v--" + str(i.arb.data) + "->" 
            i = i.next
            
        print(str1)
        
        
class Solution:
    def clone_linked_list(self,head):
        hash_map = {}
        
        curr = head
        
        while(curr):
            hash_map[curr] = Node(curr.data)
            curr = curr.next

        
        curr = head
        for i in range(len(hash_map)):
            curr = curr.next
        
        curr = head 
        
        while(curr):
            if(curr.next):
                hash_map[curr].next = hash_map[curr.next]
            hash_map[curr].arb = hash_map[curr.arb]
            curr = curr.next
        
        return hash_map[head] 
        
l1 = LinkedList()
l1.insert_data_list([1,2,3,4,5])
l1.set_arb(l1.head,1,2)
l1.set_arb(l1.head,2,4)
l1.set_arb(l1.head,3,1)
l1.set_arb(l1.head,4,1)
l1.set_arb(l1.head,5,1)
l1.print()
print("Cloned Linked list")
sol = Solution()


cloned_list = LinkedList()
cloned_list.head = sol.clone_linked_list(l1.head)

cloned_list.print()
```

```python
class Node:
    def __init__(self, data = None, next = None):
        self.next = next
        self.data = data 
        
class LikedList:
    def __init__(self):
        self.head = None
        
    def push(self, data):
        if(self.head == None):
            self.head = Node(data)
            return 
        else:
            node = Node(data)
            node.next = self.head
            self.head = node
            return 
        
    def pop(self):
        if(self.head == None):
            print("stack is empty")
            return 
        else:
            prev = self.head
            self.head = self.head.next
            prev = None
            return 
    
    def rev(self):
        curr = self.head
        prev = None
        next = None
        
        while(curr):
            next = curr.next
            curr.next = prev
            
            prev = curr
            curr = next 
            
        return prev
            
    
    def print(self):
        if(self.head == None):
            print("stack is empty")
        else:
            self.head = self.rev()
            curr = self.head
            str1 = ""
            while(curr):
                str1 = str1 + str(curr.data) + "->"
                curr = curr.next
                
            self.head = self.rev()
            print(str1)
            return 
        
    
l1 = LikedList()
l1.push(1)
l1.print()
l1.push(2)
l1.print()
l1.push(3)
l1.print()
l1.pop()
l1.push(4)
l1.print()
l1.pop()
l1.print()
l1.pop()
l1.push(10)
l1.print()
        
```
