```python
def double_helix(a, b):
    i = 0
    j = 0
    
    max_sum = 0
    sum_top = 0
    sum_bottom = 0
    
    while(i<len(a) and j<len(b)):
        if(a[i] < b[j]):
            sum_top = sum_top + a[i]
            i = i + 1
        elif(b[j] < a[i]):
            sum_bottom = sum_bottom + b[j]
            j = j + 1
        else:
            max_sum = max_sum + (a[i] + max(sum_bottom, sum_top))
            i = i + 1
            j = j + 1
            sum_top = 0
            sum_bottom = 0
    
    
    while(i < len(a)):
        sum_top = sum_top + a[i]
        i = i + 1
        
    while(j < len(b)):
        sum_bottom = sum_bottom + b[j]
        j = j + 1
        
    max_sum = max_sum + max(sum_bottom, sum_top)
    
    return (max_sum)
    
while(True):
    a = list(map(int, input().split()))[1:]
    
    if(len(a) == 0):
        break
    else:
        b = list(map(int, input().split()))[1:]
        print(double_helix(a, b))


        

        
        
    
            
```
